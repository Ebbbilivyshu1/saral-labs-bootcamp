8.3 ELECTROMAGNETIC WAVES
8.3.1 Sources of electromagnetic waves
How are electromagnetic waves produced? Neither stationary charges nor charges in uniform motion (steady currents) can be sources of electromagnetic waves. The former produces only electrostatic fields, while the latter produces magnetic fields that, however, do not vary with time. It is an important result of Maxwell's theory that accelerated charges radiate electromagnetic waves. The proof of this basic result is beyond the scope of this book, but we can accept it on the basis of rough, qualitative reasoning. Consider a charge oscillating with some frequency. (An oscillating charge is an example of accelerating charge.) This produces an oscillating electric field in space, which produces an oscillating magnetic field, which in turn, is a source of oscillating electric field, and so on. The oscillating electric and magnetic fields thus regenerate each other, so to speak, as the wave propagates through the space. The frequency of the electromagnetic wave naturally equals the frequency of oscillation of the charge. The energy associated with the propagating wave comes at the expense of the energy of the source – the accelerated charge.

8.4.1 Radio waves
Radio waves are produced by the accelerated motion of charges in conducting wires. They are used in radio and television communication systems. They are generally in the frequency range from 500 kHz to about 1000 MHz. The AM (amplitude modulated) band is from 530 kHz to 1710 kHz. Higher frequencies upto 54 MHz are used for short wave bands. TV waves range from 54 MHz to 890 MHz. The FM (frequency modulated) radio band extends from 88 MHz to 108 MHz. Cellular phones use radio waves to transmit voice communication in the ultrahigh frequency (UHF) band. How these waves are transmitted and received is described in Chapter 15.
8.4.2 Microwaves
Microwaves (short-wavelength radio waves), with frequencies in the gigahertz (GHz) range, are produced by special vacuum tubes (called klystrons, magnetrons and Gunn diodes). Due to their short wavelengths, they are suitable for the radar systems used in aircraft navigation. Radar also provides the basis for the speed guns used to time fast balls, tennis- serves, and automobiles. Microwave ovens are an interesting domestic application of these waves. In such ovens, the frequency of the microwaves is selected to match the resonant frequency of water molecules so that energy from the waves is transferred efficiently to the kinetic energy of the molecules. This raises the temperature of any food containing water.

8.4.3 Infrared waves
Infrared waves are produced by hot bodies and molecules. This band lies adjacent to the low-frequency or long-wave length end of the visible spectrum. Infrared waves are sometimes referred to as heat waves. This is because water molecules present in most materials readily absorb infrared waves (many other molecules, for example, CO2, NH,, also absorb infrared waves). After absorption, their thermal motion increases, that is, they heat up and heat their surroundings. Infrared lamps are used in physical therapy. Infrared radiation also plays an important role in maintaining the earth's warmth or average temperature through the greenhouse effect. Incoming visible light (which passes relatively easily through the atmosphere) is absorbed by the earth's surface and re- radiated as infrared (longer wavelength) radiations. This radiation is trapped by greenhouse gases such as carbon dioxide and water vapour. Infrared detectors are used in Earth satellites, both for military purposes and to observe growth of crops. Electronic devices (for example semiconductor light emitting diodes) also emit infrared and are widely used in the remote switches of household electronic systems such as TV sets, video recorders and hi-fi systems.
8.4.4 Visible rays
It is the most familiar form of electromagnetic waves. It is the part of the spectrum that is detected by the human eye. It runs from about 4 x 1014 Hz to about 7 x 1014 Hz or a wavelength range of about 700- 400 nm. Visible light emitted or reflected from objects around us provides us information about the world. Our eyes are sensitive to this range of wavelengths. Different animals are sensitive to different range of wavelengths. For example, snakes can detect infrared waves, and the 'visible' range of many insects extends well into the utraviolet.
8.4.5 Ultraviolet rays
It covers wavelengths ranging from about 4 × 107 m (400 nm) down to 6 × 10-10m (0.6 nm). Ultraviolet (UV) radiation is produced by special lamps and very hot bodies. The sun is an important source of ultraviolet light. But fortunately, most of it is absorbed in the ozone layer in the atmosphere at an altitude of about 40-50 km. UV light in large quantities has harmful effects on humans. Exposure to UV radiation induces the production of more melanin, causing tanning of the skin. UV radiation is absorbed by ordinary glass. Hence, one cannot get tanned or sunburn through glass windows.

## Advanced Version (Electromagnetic Waves)

### 1. Introduction and Symmetry in Electromagnetism

From the basic laws of electromagnetism:

- Steady electric current produces a stationary magnetic field.  
- A magnetic field changing with time gives rise to an electric field (Faraday’s law of induction).  

James Clerk Maxwell argued that the converse must also hold:  
👉 **A time‑varying electric field produces a magnetic field.**

This completes the symmetry between electric and magnetic phenomena and leads naturally to the prediction of **self‑sustaining electromagnetic waves** propagating through space.

***

### 2. Displacement Current and Ampere–Maxwell Law

Consider a capacitor being charged with a time‑dependent current \( i(t) \).  
Using Ampere’s circuital law \( \oint \vec{B} \cdot d\vec{l} = \mu_0 i \) on a loop around the wire, we get a non‑zero magnetic field at a point P outside the capacitor.

However, if we choose a **pot‑like or tiffin‑box surface** with the same boundary loop, but with its bottom between the plates, **no conduction current passes through this surface**.

So on this surface the right‑hand side of Ampere’s law gives zero, yet the left‑hand side is the same. This is a **contradiction**.

Maxwell noticed that, while the conduction current is zero between the plates, the **electric field** between the plates is changing.

For a parallel‑plate capacitor of area \( A \) and charge \( Q \), the electric field between the plates is  
\[
E = \frac{Q}{\varepsilon_0 A} \quad \Rightarrow \quad \Phi_E = EA = \frac{Q}{\varepsilon_0}.
\]
Differentiating,
\[
\frac{d\Phi_E}{dt} = \frac{dQ}{dt} \cdot \frac{1}{\varepsilon_0} = \frac{i}{\varepsilon_0}.
\]
From this, Maxwell inferred that we must add a **displacement current**  
\[
i_d = \varepsilon_0 \frac{d\Phi_E}{dt},
\]
such that the **total current** through any surface bounded by the loop is the same, regardless of the surface chosen.
Thus the **total current** is  
\[
i = i_c + i_d,
\]
where
- \( i_c \): **conduction current** (flow of real charges in conductors),  
- \( i_d \): **displacement current** (due to time‑varying electric field / electric flux).

The **generalised Ampere–Maxwell law** then becomes  
\[
\oint \vec{B} \cdot d\vec{l} = \mu_0 i_c + \mu_0 \varepsilon_0 \frac{d\Phi_E}{dt}.
\]
This ensures that the predicted magnetic field is **consistent** for any surface with the same boundary loop.
***

### 3. Maxwell’s Equations in Vacuum

Maxwell collected four integral‑form equations that describe electric and magnetic fields and their sources (charge and current densities). They are often written in vacuum as:

1. **Gauss’s law for electricity**  
   \[
   \oint \vec{E} \cdot d\vec{A} = \frac{Q}{\varepsilon_0}
   \]
   (Electric flux through a closed surface is proportional to the net charge enclosed.)

2. **Gauss’s law for magnetism**  
   \[
   \oint \vec{B} \cdot d\vec{A} = 0
   \]
   (There is no net magnetic flux through any closed surface; no isolated magnetic monopoles.)

3. **Faraday’s law of induction**  
   \[
   \oint \vec{E} \cdot d\vec{l} = -\frac{d\Phi_B}{dt}
   \]
   (A time‑varying magnetic flux induces an electric field.)

4. **Ampere–Maxwell law**  
   \[
   \oint \vec{B} \cdot d\vec{l} = \mu_0 i_c + \mu_0 \varepsilon_0 \frac{d\Phi_E}{dt}
   \]
   (Magnetic field is produced by conduction current plus displacement current.)

These four equations, together with the **Lorentz force law** \( \vec{F} = q(\vec{E} + \vec{v} \times \vec{B}) \), mathematically express all basic laws of classical electromagnetism.
***

### 4. Electromagnetic Waves: Generation and Wave Equation

Maxwell’s key prediction: **Coupled, time‑varying electric and magnetic fields can propagate as waves through space.**  
These are **electromagnetic waves**—self‑sustaining oscillations of \( \vec{E} \) and \( \vec{B} \).

Important results:

- **Sources of EM waves:**  
  - **Stationary charges** produce only **electrostatic fields**.  
  - **Steady currents** produce **static magnetic fields**.  
  - **Accelerated charges** radiate **electromagnetic waves**.

  An **oscillating charge** (e.g., harmonic motion) acts as a **basic source** of EM waves. The frequency of the wave is the same as the oscillation frequency of the charge.

- **Hertz’s experiment (1887):**  
  Hertz produced and detected **radio‑like electromagnetic waves** in vacuum, confirmed their wavelength and speed, and showed that their reflection, refraction, and polarization resemble those of light and heat waves. This experiment **verified Maxwell’s theoretical prediction**.

- **Later work:**  
  - J. C. Bose produced **shorter‑wavelength waves** (microwaves) in the laboratory.  
  - Guglielmo Marconi transmitted EM waves **over long distances**, laying the foundation for **wireless communication**.
***

### 5. Mathematical Form of Plane EM Waves

Consider a **plane electromagnetic wave** propagating in vacuum along the **z‑direction**. The electric field \( \vec{E} \) and magnetic field \( \vec{B} \) are perpendicular to each other and to the direction of propagation.
We can write:
\[
E_x = E_0 \sin(kz - \omega t), \qquad
B_y = B_0 \sin(kz - \omega t),
\]
where:
- \( k = \frac{2\pi}{\lambda} \) is the **wave number**,  
- \( \omega \) is the **angular frequency**,  
- \( \omega/k = c \) is the **wave speed** in vacuum.
From Maxwell’s equations, one can derive the wave equation and show that  
\[
c = \frac{1}{\sqrt{\mu_0 \varepsilon_0}} \approx 3 \times 10^8\ \text{m/s}.
\]
This value turns out to be **very close to the measured speed of light**, suggesting that **light is an electromagnetic wave**.

The magnitudes of \( E \) and \( B \) at any point are related by  
\[
E_0 = c B_0,
\]
so  
\[
B_0 = \frac{E_0}{c}.
\]
This relation is used in example‑type calculations, such as:
- Given \( E = 6.3\ \text{V/m} \), find \( B \):  
  \[
  B = \frac{6.3}{3 \times 10^8} = 2.1 \times 10^{-8}\ \text{T}.
  \]
- Given the direction of \( \vec{E} \) and propagation, determine the direction of \( \vec{B} \) using \( \vec{E} \times \vec{B} \) along the propagation direction.
***

### 6. Nature, Speed, and Energy of EM Waves

- **Transverse nature:**  
  In vacuum, \( \vec{E} \) and \( \vec{B} \) are:
  - **Perpendicular to each other**,
  - **Perpendicular to the direction of propagation**.

- **No medium required:**  
  EM waves can propagate through **vacuum** (or free space) because the fields themselves sustain the oscillation.
- **Speed in vacuum:**  
  \[
  c = \frac{1}{\sqrt{\mu_0 \varepsilon_0}} \approx 3 \times 10^8\ \text{m/s}.
  \]
  The constancy of \( c \) with wavelength is experimentally well established and is used to **define the standard of length**.

- **Speed in a material medium:**  
  If the medium has permittivity \( \varepsilon \) and permeability \( \mu \), the wave speed becomes  
  \[
  v = \frac{1}{\sqrt{\mu \varepsilon}}.
  \]
  This leads to the concept of **refractive index** \( n = c/v \), which is introduced in the next chapter.

- **Energy transport:**  
  EM waves **carry energy**; the energy comes at the expense of the energy of the **accelerated charge** that radiates.  
  The **average energy density** of the electric field equals the **average energy density** of the magnetic field for a harmonic plane wave:
  \[
  u_E = \frac{1}{2} \varepsilon_0 E_0^2, \quad
  u_B = \frac{B_0^2}{2\mu_0}, \quad
  \text{and} \quad
  u_E = u_B.
  \]
  This is explored in exercise‑style problems (e.g., average energy‑density equality at \( 2.0 \times 10^{10}\ \text{Hz} \), \( E_0 = 48\ \text{V/m} \)).
***

### 7. Electromagnetic Spectrum (Detailed View)

The spectrum of EM waves in principle spans an **infinite** range of wavelengths, from **gamma rays** to **long radio waves**. The classification is mainly based on **how they are produced and detected**, not on sharp boundaries.

#### Broad wavelength/frequency regions

- **Gamma rays:**  
  \(\lambda \sim 10^{-14}\ \text{m}\) to below \(10^{-12}\ \text{m}\); produced in **nuclear reactions** and by **radioactive nuclei**; used to destroy cancer cells.

- **X‑rays:**  
  \(\lambda \sim 10^{-10}\ \text{m}\) down to \(\sim 10^{-13}\ \text{m}\); produced by **high‑energy electrons** hitting a metal target; used in diagnostics and cancer treatment.
- **Ultraviolet (UV):**  
  \(\lambda \sim 4 \times 10^{-7}\ \text{m}\) to \(6 \times 10^{-10}\ \text{m}\); produced by **very hot bodies** and **special lamps**; the Sun is a major source, but most UV is absorbed in the **ozone layer**.
- **Visible light:**  
  \(\lambda \sim 700\ \text{nm}\) to \(400\ \text{nm}\); detected by the **human eye**; the Sun’s peak radiation is in this region, which matches the **evolutionary tuning** of human vision.

- **Infrared (IR):**  
  Just below visible; produced by **hot bodies** and **molecules**; often called **heat waves** because many molecules (e.g., H₂O, CO₂, NH₃) absorb it and increase their thermal energy.

- **Microwaves:**  
  Short‑wavelength radio waves, \( f \sim \text{GHz} \); produced by **klystrons, magnetrons, Gunn diodes**; used in radar, speed‑guns, and microwave ovens (where frequency is tuned to water‑molecule resonance).

- **Radio waves:**  
  Longest wavelength EM waves, up to \(\lambda \sim 10^6\ \text{m}\); produced by **accelerated electrons in conducting wires / antennas**; used in **radio, TV, FM, AM, mobile communication, and satellite systems**.

#### Photon‑energy perspective (optional but relevant)

Using the **photon energy** formula \( E = h\nu \) (from quantum ideas), one can compute the energy (in eV) across the spectrum.  
This exercise shows:
- **High‑frequency** waves (gamma, X‑rays, UV) carry **high‑energy photons**, which can ionize atoms and damage tissue.  
- **Low‑frequency** waves (radio, microwaves, IR) carry **low‑energy photons**, which mainly cause heating or collective motion of charges and molecules.
***

### 8. Interaction with Matter and System Size

The way EM waves interact with matter depends strongly on **frequency** (or wavelength) and on the **nature of atoms and molecules** in the medium.
Key qualitative ideas:

- **Infrared waves:**  
  Vibrations of **entire atoms or molecules** are excited; this increases internal energy and temperature, hence the term **heat waves**.

- **Visible and UV:**  
  Cause **electronic transitions** (jump between energy levels) in atoms. UV can also **ionize** atoms and is strongly absorbed by glass and ozone.
- **X‑rays:**  
  Interact with **inner‑shell electrons** and can penetrate soft tissue but are absorbed by bone, making them useful for imaging.

- **Radio and microwaves:**  
  Typically interact via **collective oscillation of free or bound charges** in conductors and dielectrics.
A useful rule of thumb:
- The **wavelength** of EM radiation is often comparable to the **characteristic size** of the radiating system.  
  - **Gamma rays** (nuclear size) \(\sim 10^{-14}\ \text{m}\).  
  - **X‑rays** from heavy atoms.  
  - **Radio waves** from antennas roughly of the order of the wavelength.  
  - **Visible light** from atoms, but wavelength is **much larger** than the atom size.
***

### 9. Key Conclusions and Summary of Physics

- **Displacement current** \( i_d = \varepsilon_0 d\Phi_E/dt \) resolves the inconsistency in Ampere’s law and ensures that the **total current** is well‑defined for any surface with a given boundary.
- **Maxwell’s equations** predict that:
  - **Time‑varying electric and magnetic fields** can mutually sustain each other.  
  - The resulting **electromagnetic waves** propagate in vacuum at speed \( c = 1/\sqrt{\mu_0 \varepsilon_0} \).  
  - **Light** is one such electromagnetic wave.

- **Accelerated charges** (especially oscillating charges) are the **fundamental sources** of EM radiation; the frequency of the wave equals the oscillation frequency of the source.

- The **entire electromagnetic spectrum**—from gamma rays to radio waves—consists of the same underlying physical phenomenon, differing only in **wavelength/frequency** and hence in **ways of production, detection, and interaction with matter**.

- EM waves **carry energy and momentum**, enabling:
  - **Radio and TV broadcasting**,  
  - **Microwave communication and radar**,  
  - **Medical imaging and therapy**,  
  - **Solar energy transfer** to Earth.
